Question 1:  What is Simple Linear Regression?

ANS> Simple linear regression models the relationship between a dependent variable and a single independent variable. It tries to provide best fit line so that the sum of squared of residual(error) wil be minimum.
The relationship between the dependent and independent variables is represented by the simple linear equation:

y=mx+b

Here:

* y is the predicted value (dependent variable).
* m is the slope of the line
* x is the independent variable.
* b is the y-intercept (the value of y when x is 0).
In this equation m signifies the slope of the line indicating how much y changes for a one-unit increase in x, a positive m suggests a direct relationship while a negative m indicates an inverse relationship.

Question 2: What are the key assumptions of Simple Linear Regression?

ANS>
1. **Linearity:**
Assumption: The relationship between the independent and dependent variables is linear.The first and foremost assumption of linear regression is that the relationship between the predictor(s) and the response variable is linear. This means that a change in the independent variable results in a proportional change in the dependent variable.
2. **Homoscedasticity of Residuals in Linear Regression:**  
Homoscedasticity is one of the key assumptions of linear regression, which asserts that the residuals (the differences between observed and predicted values) should have a constant variance across all levels of the independent variable(s). In simpler terms, it means that the spread of the errors should be relatively uniform, regardless of the value of the predictor.
3. ** Multivariate Normality - Normal Distribution: **
Multivariate normality is a key assumption for linear regression models when making statistical inferences. Specifically, it means that the residuals (the differences between observed and predicted values) should follow a normal distribution when considering multiple predictors together.
4. **Independence of Errors:**
Independence of errors is another critical assumption for linear regression models. It ensures that the residuals (the differences between the observed and predicted values) are not correlated with one another. This means that the error associated with one observation should not influence the error of any other observation.
5. **Lack of Multicollinearity:**
Assumption: The independent variables are not highly correlated with each other.Multicollinearity occurs when two or more independent variables in the model are highly correlated, leading to redundancy in the information they provide. This can inflate the standard errors of the coefficients, making it difficult to determine the effect of each independent variable.
6.  **Absence of Endogeneity:**
No endogeneity is an important assumption in the context of multiple linear regression. The assumption of no endogeneity states that the independent variables in the regression model should not be correlated with the error term.

Question 3: What is heteroscedasticity, and why is it important to address in regression models?

ANS> There are certain assumptions that is important to folow while performing regression, that is **homoscedasticity**: the spread of the errors should be relatively uniform, therefore violation of this assumption is called Heteroscedasticity.

* It is important to address in regression models because it increases the variance of the regression coefficient estimates, making the results of the analysis unreliable. This can lead to incorrect conclusions about the statistical significance of the terms in the model.
* Additionally, the OLS(Ordinary Least Square) estimators are no longer the Best Linear Unbiased Estimator (BLUE), and hypothesis tests like the t-test and F-test become invalid due to inconsistencies in the covariance matrix of the estimated regression coefficients.

Question 5: What is polynomial regression, and how does it differ from linear
regression?

ANS>  s an extension of linear regression that models the relationship between the dependent variable and the independent variable(s) as an n-th degree polynomial. The equation for polynomial regression.
* It can model non-linear relationships by fitting a polynomial equation to the data. Ideal for complex patterns.
* Offers greater flexibility, can model curves and intricate patterns, suitable for non-linear trends.

```python
# Question 6:  Implement a Python program to fit a Simple Linear Regression model to
# the following sample data:
# ● X = [1, 2, 3, 4, 5]
# ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
# Plot the regression line over the data points.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = [1, 2, 3, 4, 5]
Y = [2.1, 4.3, 6.1, 7.9, 10.2]

df = pd.DataFrame({'X': X, 'Y': Y})

model = LinearRegression()
model.fit(df[['X']], df['Y'])

plt.scatter(df['X'], df['Y'], label='Data Points')
plt.plot(df['X'], model.predict(df[['X']]), color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Question 7: Fit a Multiple Linear Regression model on this sample data:
# ● Area = [1200, 1500, 1800, 2000]
# ● Rooms = [2, 3, 3, 4]
# ● Price = [250000, 300000, 320000, 370000]
# Check for multicollinearity using VIF and report the results.

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

df = pd.DataFrame({'Area': Area, 'Rooms': Rooms, 'Price': Price})

model = LinearRegression()
data1 = model.fit(df[['Area', 'Rooms']], df['Price'])

plt.scatter(df['Area'], df['Price'], label='Data Points')
plt.plot(df['Area'], model.predict(df[['Area', 'Rooms']]), color='red', label='Regression Line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()

# Calculate VIF for each independent variable
vif_data = pd.DataFrame()
vif_data["feature"] = df[['Area', 'Rooms']].columns
vif_data["VIF"] = [variance_inflation_factor(df[['Area', 'Rooms']].values, i)
                   for i in range(len(df[['Area', 'Rooms']].columns))]

print("VIF Results:")
display(vif_data)

print("Reports: Both 'Area' and 'Rooms' have a VIF of approximately 127.8. This is a very high value")
print("suggesting significant multicollinearity between these two variables.")
print("This means that 'Area' and 'Rooms' are highly correlated with each other,\nwhich can make it difficult for the model to determine the individual impact of each variable on the 'Price'.")
```

```python
# Question 8:  Implement polynomial regression on the following data:
# ● X = [1, 2, 3, 4, 5]

# ● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
# Fit a 2nd-degree polynomial and plot the resulting curve.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


X = [1, 2, 3, 4, 5]
Y = [2.2, 4.8, 7.5, 11.2, 14.7]

df = pd.DataFrame({'X': X, 'Y': Y})

poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(df[['X']])

model = LinearRegression()
model.fit(X_poly, df['Y'])

plt.scatter(df['X'], df['Y'], label='Data Points')
plt.plot(df['X'], model.predict(X_poly), color='red', label='Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Question 9: Create a residuals plot for a regression model trained on this data:
# ● X = [10, 20, 30, 40, 50]
# ● Y = [15, 35, 40, 50, 65]
# Assess heteroscedasticity by examining the spread of residuals.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = [10, 20, 30, 40, 50]
Y = [15, 35, 40, 50, 65]

df = pd.DataFrame({'X': X, 'Y': Y})

model = LinearRegression()
model.fit(df[['X']], df['Y'])

# Calculate residuals
residuals = df['Y'] - model.predict(df[['X']]) # df['Y'] is actual values

# Create residuals plot
plt.scatter(df['X'], residuals)
plt.axhline(y=0, color='red', linestyle='--') # Add a horizontal line at y=0
plt.xlabel('X')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()

# Assess heteroscedasticity
print("To assess heteroscedasticity, examine the spread of the residuals in the plot.")
print("according to the plot, there might be some degree of heteroscedasticity in the model")
```

Question 10: Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.

ANS>  
1. Steps to Address Heteroscedasticity
* **Visual Diagnostics:** Plot residuals vs. predicted values or each predictor to confirm the pattern of changing variance.
* **Transform Dependent Variable:** Apply transformations such as:
Logarithm of house price (log(price)) to stabilize variance.
* **Robust Standard Errors:** Use heteroscedasticity-consistent (e.g., White’s) standard errors to obtain valid inference without transforming the data.
* **Weighted Least Squares (WLS):** Assign weights inversely proportional to the variance of observations, so observations with larger variance get lower weight.
* **Generalized Least Squares (GLS):** Model variance structure explicitly if known or approximated.

2. Steps to Address Multicollinearity
* **Correlation Matrix & VIF Analysis:** Quantify multicollinearity via variance inflation factor (VIF). VIF > 5 or 10 indicates problematic multicollinearity.
* **Remove or Combine Correlated Variables:**
Drop redundant predictors.
Create composite variables or indices through domain knowledge (e.g., total living area instead of separate room counts and area features).
* **Dimensionality Reduction:**
Apply Principal Component Analysis (PCA) or Partial Least Squares (PLS) to transform correlated predictors into orthogonal components.
* **Regularization Techniques:**
Use regression models like Ridge Regression (L2 regularization) which shrinks correlated coefficients and reduces variance.

Model Comparison: Compare adjusted R², AIC/BIC, or out-of-sample prediction error to select the best modeling approach.
